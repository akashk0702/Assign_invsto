# check for outliers

```python
# Function to detect outliers using IQR
def detect_outliers_iqr(ohlc_data):
    outliers = pd.DataFrame(columns=ohlc_data.columns)
    for col in ohlc_data.columns:
        q1 = ohlc_data[col].quantile(0.25)
        q3 = ohlc_data[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        col_outliers = ohlc_data[(ohlc_data[col] < lower_bound) | (ohlc_data[col] > upper_bound)]
        outliers = pd.concat([outliers, col_outliers], axis=0)
    return outliers

# Detect outliers
outliers_ohlc_data = detect_outliers_iqr(ohlc_data)
print("Outliers detected using IQR:")
print(outliers_ohlc_data)
```

# capped

```python

# Function to cap outliers using IQR
def cap_outliers_iqr(df):
    capped_df = df.copy()
    for col in df.columns:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        capped_df[col] = capped_df[col].apply(lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x)
    return capped_df

```

# find the technical indicators

```python
def calculate_technical_indicators(data_new_sorted, window_size=20):
    # Calculate Moving Averages
    data_new_sorted['SMA'] = data_new_sorted['Close'].rolling(window=window_size).mean()
    
    # Calculate Bollinger Bands
    data_new_sorted['std'] = data_new_sorted['Close'].rolling(window=window_size).std()
    data_new_sorted['upper_band'] = data_new_sorted['SMA'] + 2 * data_new_sorted['std']
    data_new_sorted['lower_band'] = data_new_sorted['SMA'] - 2 * data_new_sorted['std']
    
    # Calculate Relative Strength Index (RSI)
    delta = data_new_sorted['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window_size).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window_size).mean()
    rs = gain / loss
    data_new_sorted['RSI'] = 100 - (100 / (1 + rs))
    
    return data_new_sorted


# Calculate technical indicators
data_new_sorted = calculate_technical_indicators(data_new_sorted)

# # Print DataFrame with technical indicators
# print(data_new_sorted)
```

```python

# Calculate volatility measures
data_new_sorted['Daily_Return'] = data_new_sorted['Close'].pct_change()  # Daily returns
data_new_sorted['Volatility_7D'] = data_new_sorted['Daily_Return'].rolling(window=7).std()  # 7-day rolling standard deviation of returns
data_new_sorted['Volatility_30D'] = data_new_sorted['Daily_Return'].rolling(window=30).std()  # 30-day rolling standard deviation of returns

# Price patterns
data_new_sorted['SMA_50'] = data_new_sorted['Close'].rolling(window=50).mean()  # 50-day simple moving average
data_new_sorted['SMA_200'] = data_new_sorted['Close'].rolling(window=200).mean()  # 200-day simple moving average

# Relative Strength Index (RSI)
def calculate_rsi(series, window=14):
    delta = series.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

data_new_sorted['RSI'] = calculate_rsi(data_new_sorted['Close'])

# Moving Average Convergence Divergence (MACD)
short_window = 12
long_window = 26
signal_window = 9

ema_short = data_new_sorted['Close'].ewm(span=short_window, min_periods=1).mean()
ema_long = data_new_sorted['Close'].ewm(span=long_window, min_periods=1).mean()

macd = ema_short - ema_long
signal_line = macd.ewm(span=signal_window, min_periods=1).mean()

data_new_sorted['MACD'] = macd
data_new_sorted['Signal_Line'] = signal_line

# Drop NaN values resulting from rolling calculations
data_new_sorted.dropna(inplace=True,ignore_index=True)

# Now you have a DataFrame with new features that can be used for your trading strategy
print(data_new_sorted.head())
```

```python
import sqlite3
import pandas as pd

# Function to create the OHLC table
def create_ohlc_table(conn):
    c = conn.cursor()
    c.execute('''CREATE TABLE IF NOT EXISTS ohlc (
                 date TEXT,
                 open REAL,
                 high REAL,
                 low REAL,
                 close REAL,
                 year INTEGER,
                 month INTEGER,
                 PRIMARY KEY (date)
                 )''')
    conn.commit()

# # Function to insert OHLC data into the table
# def insert_ohlc_data(conn, df):
#     c = conn.cursor()
#     for index, row in df.iterrows():
#         date = row['Date']
#         open_price = row['Open']
#         high_price = row['High']
#         low_price = row['Low']
#         close_price = row['Close']
#         year = int(date.split('-')[0])
#         month = int(date.split('-')[1])
#         c.execute("INSERT OR REPLACE INTO ohlc VALUES (?, ?, ?, ?, ?, ?, ?)",
#                   (date, open_price, high_price, low_price, close_price, year, month))
#     conn.commit()

# Function to insert OHLC data into the table
def insert_ohlc_data(conn, df):
    c = conn.cursor()
    for index, row in df.iterrows():
        date = row['Date'].strftime('%Y-%m-%d')  # Convert Timestamp to string in 'YYYY-MM-DD' format
        open_price = row['Open']
        high_price = row['High']
        low_price = row['Low']
        close_price = row['Close']
        year = int(date.split('-')[0])
        month = int(date.split('-')[1])
        c.execute("INSERT OR REPLACE INTO ohlc VALUES (?, ?, ?, ?, ?, ?, ?)",
                  (date, open_price, high_price, low_price, close_price, year, month))
    conn.commit()


# Function to query OHLC data by year and month
def query_ohlc_data(conn, year, month):
    c = conn.cursor()
    c.execute("SELECT * FROM ohlc WHERE year=? AND month=?", (year, month))
    data = c.fetchall()
    return pd.DataFrame(data, columns=['Date', 'Open', 'High', 'Low', 'Close', 'Year', 'Month'])

# Connect to the SQLite database (creates a new one if not exists)
conn = sqlite3.connect('ohlc_data.db')

ohlc_data = data_new_sorted
# Create the OHLC table if not exists
create_ohlc_table(conn)

# Insert OHLC data into the table
insert_ohlc_data(conn, ohlc_data)

# Query OHLC data for a specific year and month
query_result = query_ohlc_data(conn, 2024, 1)
print("Query Result:")
print(query_result)

# Close the database connection
conn.close()
```
